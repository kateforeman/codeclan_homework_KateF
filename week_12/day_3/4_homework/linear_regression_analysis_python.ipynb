## Linear Regression Building Homework 

We want you to build a linear regression model to predict the selling_price of a car in both in Python and in R.  

Do some basic data exploration in Python (you may wish to use pandas_profiling for this if it works for you)  


```python
import pandas as pd 

cars = pd.read_csv("car_dekho_details.csv") 
```

```python
cars 
cars = cars.drop('name', 1) 
```

```python
import pandas_profiling 
```

```python
#pandas_profiling.ProfileReport(cars) 
```

```python
cars['year'] = cars['year'].astype('category') 
```

```python
cars = pd.get_dummies(cars, drop_first = True) 
```

```python
from sklearn.linear_model import LinearRegression 
```

```python
y = cars["selling_price"]  
x = cars.drop(columns = "selling_price") 
```

```python
model = LinearRegression() 
```

```python
model.fit(x, y) 
```

```python
model.score(x, y) 
```

```python
model.intercept_ 
```

```python
model.coef_ 
```

```python
pd.DataFrame({
    "Variables" : x.columns.values, 
    "Coefficients" : model.coef_}) 
```

```python
import statsmodels.api as sm 
```

```python
x_i = sm.add_constant(x) 
x_i 
```

```python
sm_model = sm.OLS(y, x_i).fit() 
print(sm_model.summary()) 
```

**Interpretation:** The r-squared is exactly the same as the value I got in my model built in R. It is a small r2 of 0.485, indicating that the model can only predict 49% of the variance of selling price of cars.
